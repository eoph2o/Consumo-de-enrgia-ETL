Este projeto tem como objetivo demonstrar uma aplicação simples de ETL (Extract, Transform, Load).
A partir de uma base de dados contendo o consumo mensal de energia elétrica, realizou-se a extração dos valores de consumo em kWh e das correntes elétricas correspondentes a cada mês.
Em seguida, esses dados foram transformados por meio do cálculo das médias mensais de consumo. Por fim, os resultados obtidos foram carregados em um novo arquivo, no qual é apresentada a relação das médias mensais calculadas.

```python
import pandas as pd
janeiro = pd.read_csv('jan.csv')
fevereiro = pd.read_csv('fev.csv')
marco = pd.read_csv('mar.csv')
abril = pd.read_csv('abr.csv')
maio = pd.read_csv('mai.csv')
junho = pd.read_csv('jun.csv')
meses = [janeiro, fevereiro, marco, abril, maio, junho]

#print(meses)
listaDeMediasDeConsumo = []
listaDeMediasDeCorrente = []
listaDeMediasDeTensao = []
for consumo in meses:
 listaDeConsumo =  consumo["consumo_kWh"].tolist()
 #print(listaDeConsumo)
 media = sum(listaDeConsumo)/len(listaDeConsumo)
 #print(f"{media:.1f}")
 listaDeMediasDeConsumo.append(round(media, 1))
print(f"media das kWh mensal {listaDeMediasDeConsumo}")

for consumo in meses:
 listaDeConsumo =  consumo["corrente_A"].tolist()
 #print(listaDeConsumo)
 media = sum(listaDeConsumo)/len(listaDeConsumo)
 #print(f"{media:.1f}")
 listaDeMediasDeCorrente.append(round(media, 1))
print(f"media das correntes mensal {listaDeMediasDeCorrente}")

```

```python
df = pd.read_csv('usoMensalDeEnergia.csv')
df["kWhMensal"]= listaDeMediasDeConsumo
df["correnteMensal"]= listaDeMediasDeCorrente
df.to_csv('usoMensalDeEnergia.csv', index=False)
```
